```python
#   Kyle Territo   #
#   CHE 7140       #
#   HP 03_06       #

import numpy as np
import matplotlib.pyplot as plt
import math

#given
CA0 = 3.6
H = -20750
Cp = 125
UA = 5*10**6
VR = 800
Tc = 436

#step size and initial conditions
Dt = 0.001
t0 = 0
tf = 0.4
f0 = 0
Temp0 = 426

n_steps = int(round((tf-t0)/Dt))
```

```python
# Creating arrays for f, Temp, and t
f_array = np.zeros(n_steps + 1)
Temp_array = np.zeros(n_steps + 1)
t_array = np.zeros(n_steps + 1)

f_array[0]=f0
Temp_array[0]=Temp0
t_array[0]=t0
```

```python
# Creating a For-loop that iterates both ODEs and stores iteration values
for i in range (1, n_steps + 1):
  # Setting variables   
    f = f_array[i-1]
    Temp = Temp_array[i-1]
    t = t_array[i-1]
  # ODEs
    dfdt = ((2.61*10**14)*math.exp(-14570/Temp)*CA0*(1-f))/CA0
    dTempdt = ((UA*(Tc-Temp))+(-H*((2.61*10**14)*math.exp(-14570/Temp))*CA0*(1-f)*VR))/(CA0*Cp*VR)
  # Storing iteration values
    f_array[i] = f + Dt*dfdt
    Temp_array[i] = Temp + Dt*dTempdt
    t_array[i] = t +Dt
```

```python
# Plotting f vs. t
fig = plt.figure()     
plt.plot(t_array, f_array, linewidth = 1, label = 'f')
plt.title('f vs t', fontsize = 12)    
plt.xlabel('t (hr)', fontsize = 12)
plt.ylabel('f', fontsize = 12)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.grid(True)                      
plt.legend()
plt.show()
```

```python
# Plotting Temp vs. t
fig = plt.figure()   
plt.plot(t_array, Temp_array, linewidth = 1, label = 'Temp')
plt.title('Temp vs t', fontsize = 12)    
plt.xlabel('t (hr)', fontsize = 12)
plt.ylabel('Temp', fontsize = 12)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.grid(True)                      
plt.legend()
plt.show()
```
